## Exercise 3: The Quadratic Solver

In Machine Learning, we often look for the "minimum" of a curve. Many of these curves are defined by **Quadratic Equations**: $ax^2 + bx + c = 0$. Your task is to build a solver that finds the roots ($x$) for any given coefficients $a, b,$ and $c$.

**Task:**
1. Takes three inputs: `a`, `b`, and `c` (convert them to `float`).
2. Calculates the discriminant `D`.
3. Uses `if/elif/else` to print the number of roots and their values.
4. *Formula for roots:* $x = \frac{-b \pm \sqrt{D}}{2a}$

**Expected Output:**
```
Enter a: 1
Enter b: -5
Enter c: 6

Discriminant: 1.0
Two real roots found: x1 = 3.0, x2 = 2.0
```

````python
print("Quadratic Equation Solver")

a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

delta = b**2 - 4*a*c
print(f"Discriminant = {delta}")

if delta > 0:
    root1 = (-b + delta**0.5) / (2*a)
    root2 = (-b - delta**0.5) / (2*a)
    print(f"The equation has two real roots: x1 =  {root1:.2f}, x2 = {root2:.2f}")

elif delta == 0:
    root = -b / (2*a)
    print(f"The equation has one real root: x1 = x2 = {root:.2f}")

else:
    real_part = -b / (2*a)
    imaginary_part = (abs(delta)**0.5) / (2*a)
    
    print(f"The equation has two complex roots:")
    print(f"x1: {real_part:.2f} + {imaginary_part:.2f}i")
    print(f"x2: {real_part:.2f} - {imaginary_part:.2f}i")
````

## Exercise 9: Data Cleaning

In Data Science, datasets often have "dirty" data (None or empty strings) that crash AI models.


**Task**: Given the list: `raw_data = [15.5, None, 20.0, "Empty", 18.2, 10.0]`


1.Create a new list called `cleaned_data`.

2.Loop through `raw_data` and only add numbers `(float or int)` to the new list.

3.Calculate and print the average of the `cleaned_data`.

**Expected Output**
```
Cleaned Data: [15.5, 20.0, 18.2, 10.0]
The average of the cleaned data is: 15.925
```

````python
raw_data = [15.5, None, 20.0, "Empty", 18.2, 10.0]


cleaned_data = []


for item in raw_data:
    if isinstance(item, (int, float)):
        cleaned_data.append(item)

if len(cleaned_data) > 0:
    average = sum(cleaned_data) / len(cleaned_data)
    print(f"Cleaned Data: {cleaned_data}")
    print(f"The average of the cleaned data is: {average}")
````

````python
raw_data = [15.5, None, 20.0, "Empty", 18.2, 10.0]

cleaned_data = [x for x in raw_data if isinstance(x, (int, float))]

if cleaned_data:
    average = sum(cleaned_data) / len(cleaned_data)
    print(f"Cleaned Data: {cleaned_data}")
    print(f"Average: {average:.2f}")
````

## Exercise 10: Word Frequency

This is a classic CS50/MIT problem. you will learn how to use Dictionaries to count occurrences.


**Task**: Given the string: `text = "data science is data and ai is ai"`


1.Split the string into a list of words using `.split()`.

2.Create an empty dictionary called `word_counts`.

3.Loop through the words. If a word is already in the dictionary, increase its value by 1. If not, add it with a value of 1.

4.Print the final dictionary

**Expected Output**

```Word Counts: {'data': 2, 'science': 1, 'is': 2, 'and': 1, 'ai': 2}```


````python
text = "data science is data and ai is ai"

#Split the string into a list of words
words = text.split()

#Create an empty dictionary
word_counts = {}

#Loop through the words and count occurrences
for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

print(f"Word Counts: {word_counts}")
````

````python
text = "data science is data and ai is ai"
word_counts = {}

for word in text.split():
    
    word_counts[word] = word_counts.get(word, 0) + 1

print(f"Word Counts: {word_counts}")
````

## Exercise 11: The Weather Station

You are developing software for an IoT Weather Station. The hardware records temperatures in **Fahrenheit**, but your AI model requires the data in **Celsius** for global standardization. You need to build a modular system to convert and analyze this data.

**Task 1: The Converter Function**
Define a function called `to_celsius(f)` that:
1. Takes a Fahrenheit value as an input parameter.
2. Performs the calculation: $C = (F - 32) \times \frac{5}{9}$
3. **Returns** the calculated Celsius value.

**Task 2: The Analysis Function**
Define a function called `analyze_weather(temp_list)` that:
1. Iterates through a list of Fahrenheit temperatures.
2. Calls your `to_celsius()` function inside the loop for each value.
3. Prints a message for each temperature: `"Safe"` if below 35°C, or `"Heat Warning"` if 35°C or higher.

**Task 3: Execution**
Create a list of test data: `temps = [72, 95, 105, 60]` and pass this list as an argument to your `analyze_weather()` function.

**Expected Output:**
```
Temp: 22.22C - Safe
Temp: 35.00C - Heat Warning
Temp: 40.56C - Heat Warning
Temp: 15.56C - Safe
```

````python
# TO DO 
def to_celsius(f):
    # Your code 
    c = (f - 32) * 5/9
    return c

def analyze_weather(temp_list):
    # Your code 
    for f_temp in temp_list:
        c_temp = to_celsius(f_temp)

        if c_temp < 35:
            status = "Safe"
        else:
            status = "Heat Warning"
        print(f"Temp: {c_temp:.2f}°C - {status}")

# Test data
temps = [72, 95, 105, 60]
analyze_weather(temps)
````

## Exercise 12: Area Calculators

You are building a module for a **Computer Vision AI** that needs to calculate the area of different shapes detected in an image. Instead of writing one giant block of code, you need to build a **modular system** where each shape has its own specialized function. This makes the system easier to test and integrate into larger AI pipelines.


### Task 1:
Define three separate functions to handle the geometry math. Each function must **return** the calculated value.

1.  **`area_circle(radius)`**: Returns $\pi r^2$ (Use `math.pi` from the math module).
2.  **`area_rectangle(width, height)`**: Returns $width \times height$.
3.  **`area_triangle(base, height)`**: Returns $0.5 \times base \times height$.


### Task 2:
Create a main "controller" function called **`calculate_shape_area()`**. Inside this function:

1.  **User Input:** Ask the user which shape they want to calculate (*Circle*, *Rectangle*, or *Triangle*).
2.  **Branching Logic:** Use `if-elif-else` to determine which shape was chosen.
3.  **Dimension Inputs:** Inside each branch, ask the user for the specific dimensions needed (e.g., if they chose Rectangle, ask for width and height).
4.  **Function Call:** Call the corresponding function from Task 1 using the user's inputs.
5.  **Output:** Print the result formatted to 2 decimal places.

````python
import math

def area_circle(radius):
    return math.pi*radius**2

def area_rectangle(width, height):
    return width * height 

def area_triangle(base, height):
    return 0.5 * base * height

def calculate_area():
    choices = input("Enter shape(circe, rectangle, triangle): ").strip().lower()

    if choices == 'circle':
        radius = float(input("Enter radius:"))
        result = area_circle(radius)
        print((f" Area of circle is : {result:.2f} "))
    
    if choices == 'rectangle':
        w = float(input("Enter width:"))
        h = float(input("Enter height:"))
        result = area_rectangle(w,h)
        print((f" Area of rectangle is : {result:.2f} "))
    
    if choices == 'triangle':
        base = float(input("Enter base:"))
        height = float(input("Enter height:"))
        result = area_triangle(base, height)
        print((f" Area of triangle is : {result:.2f} "))
````

````python
# TO DO

# Task 1:
def cal_mean(data):
    return sum(data)/len(data)

def cal_min(data):
    return min(data)

def cal_max(data):
    return max(data)

def cal_range(data):
    return cal_max(data)-cal_min(data)

def cal_median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    mid = n // 2

    if n % 2 == 0:
        return (sorted_data[mid-1] + sorted_data[mid])/2
    else:
        return sorted_data[mid]

# Task 2: 
def generate_report(data_list):
    stats = {
        "mean":cal_mean(data_list),
        'min' : cal_min(data_list),
        'max' : cal_max(data_list),
        'range' : cal_range(data_list),
        'median' : cal_median(data_list)
    }
    
    print(f" {'Statistics':<15} | {'Value':<10} ")
    print("-"*28)

    for key, value in stats.items():
        print((f"{key:<15} | {value:<10} "))
````

````python
# Test data:
experiment_results = [10, 20, 30, 40, 50]
generate_report(experiment_results)
````
